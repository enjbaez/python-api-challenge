```python
# import dependencies

import pandas as pd
import numpy as np
import pandas as pd
import requests
import json
import seaborn as sns
import matplotlib.pyplot as plt
from citipy import citipy
%matplotlib inline
```

```python
# random list for latitudes and longitudes 

latitude = []
longitude = []

for each_lat in np.random.randint(-90,90,2000):
    latitude.append(each_lat)

for each_lon in np.random.randint(-180,180,2000):
    longitude.append(each_lon)
```

```python
# random lat and lon storage

latitude_and_Longitude = tuple(zip(latitude,longitude))
```

```python
# find the nearest city for all latitudes and longitudes

cities = []
country = []

for lat,lon in latitude_and_Longitude:
    city = citipy.nearest_city(lat,lon)
    
    cityName = city.city_name
    
    cities.append(cityName)
    
    country.append(city.country_code)
```

```python
# cities generate

print('{} different cities listed.'.format(len(set(cities))))
```

```python
# cities and countries df

countries_df = pd.DataFrame({'Cities': cities,
              'Country': country})

#add additional blank columns to store information from openweathermap api

countries_df['Latitude'] = ''
countries_df['Longitude'] = ''
countries_df['Temperature (F)'] = ''
countries_df['Humidity (%)'] = ''
countries_df['Cloudiness (%)'] = ''
countries_df['Wind Speed (mph)'] = ''

countries_df = countries_df.drop_duplicates(subset=['Cities'],keep='first')
```

```python
api_key = "29af9421ee1d1f65ef06c3b7ad9f6964"
```

```python
#loop through all rows and fill in values for blank columns in dataframe

for index,row in countries_df.iterrows():
    city_name = row['Cities']
    Country_id = row['Country']
    
    
    #api call for all Cities 
    api_url = "http://api.openweathermap.org/data/2.5/forecast" \
    "?q={},{}&units=IMPERIAL&mode=json&APPID={}".format(city_name,Country_id,api_key)
    
    
    country_info = requests.get(api_url).json()
    
    # api for each city
    print(api_url)
    
    
    try:
        countries_df.set_value(index,'Latitude',country_info['city']['coord']['lat'])
        countries_df.set_value(index,'Longitude',country_info['city']['coord']['lon'])
        countries_df.set_value(index,'Temperature (F)',country_info['list'][0]['main']['temp'])
        countries_df.set_value(index,'Humidity (%)',country_info['list'][0]['main']['humidity'])
        countries_df.set_value(index,'Cloudiness (%)',country_info['list'][0]['clouds']['all'])
        countries_df.set_value(index,'Wind Speed (mph)',country_info['list'][0]['wind']['speed'])
        
    except KeyError:
        countries_df.set_value(index,'Latitude',np.nan)
        countries_df.set_value(index,'Longitude',np.nan)
        countries_df.set_value(index,'Temperature (F)',np.nan)
        countries_df.set_value(index,'Humidity (%)',np.nan)
        countries_df.set_value(index,'Cloudiness (%)',np.nan)
        countries_df.set_value(index,'Wind Speed (mph)',np.nan)
        
        print('weather information missing')
```

```python
# numerical data update from openweathermap api
countries_df['Latitude'] = pd.to_numeric(countries_df['Latitude'])
countries_df['Longitude'] = pd.to_numeric(countries_df['Longitude'])
countries_df['Temperature (F)'] = pd.to_numeric(countries_df['Temperature (F)'])
countries_df['Humidity (%)'] = pd.to_numeric(countries_df['Humidity (%)'])
countries_df['Cloudiness (%)'] = pd.to_numeric(countries_df['Cloudiness (%)'])
countries_df['Wind Speed (mph)'] = pd.to_numeric(countries_df['Wind Speed (mph)'])

countries_df.head()
```

```python
length_of_original_DF = len(countries_df)
length_after_dropna = len(countries_df.dropna())

missing_weather_info = length_of_original_DF - length_after_dropna
```

```python
print("The following {} cities did not contain weather and will be dropped"\
      .format(missing_weather_info))

print("There are {} nonrepetitive cities. "\
      .format(len(countries_df.dropna())))
```

```python
countries_df = countries_df.dropna()
countries_df.to_csv(path_or_buf='countries_df.csv')
```
